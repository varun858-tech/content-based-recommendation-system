```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv(
    r"C:\Users\Varun Bahuguna\OneDrive\Desktop\content-based-recommendation-system\data\movies.dat",
    sep="::",
    engine="python",
    names=["id", "title", "genres"]
)
```

```python
movies["genres"] = movies["genres"].str.lower()
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x:x.split('|'))
```

```python
movies
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer()

genre_matrix = mlb.fit_transform(movies["genres"])
```

```python
genre_matrix[0]
```

```python
print(mlb.classes_)
```

```python
genre_df = pd.DataFrame(genre_matrix, columns=mlb.classes_) # convert to DataFrame
```

```python
genre_df
```

```python
genre_df.drop(columns=["(no genres listed)"], inplace=True)
```

```python
genre_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(genre_df)
```

```python
similarity
```

```python
similarity[0]
```

```python
def recommend(movie_title):
    # get index
    idx = movies[movies["title"] == movie_title].index[0]
    
    dist = similarity[idx]

    scores = sorted(list(enumerate(dist)), reverse = True, key = lambda x : x[1])[1:6]

    rec = []
    
    for i in scores:
        rec.append(movies.iloc[i[0]].title)

    return rec
```

```python
recommend("Shawshank Redemption, The (1994)")
```

```python
recommend("Fight Club (1999)")
```

```python
recommend("Iron Man (2008)")
```
