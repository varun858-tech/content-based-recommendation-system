```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = tfidf.fit_transform(movies["bag"])
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(movie_name):
    movie_name = movie_name.lower()
    
    matches = movies[movies["title"].str.lower().str.contains(movie_name)]
    
    if matches.empty:
        print("Movie not found 😅")
        return
    
    idx = matches.index[0]
    
    # 🔥 KEY LINE (on-the-fly similarity)
    dist = cosine_similarity(vectors[idx], vectors).flatten()
    
    scores = sorted(list(enumerate(dist)), reverse = True, key = lambda x : x[1])[1:6]

    rec = []
    # top 5
    for i in scores:
        rec.append(movies.iloc[i[0]].title)

    return rec
```

```python
recommend("shawshank")
```

```python
recommend("fight club")
```

```python
recommend("toy story")
```
