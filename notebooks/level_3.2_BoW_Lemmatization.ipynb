```python
import nltk
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def lemmatize_text(text):
    words = text.split()
    tagged_words = pos_tag(words)
    
    # lemmatized = [
    #     lemmatizer.lemmatize(word, get_wordnet_pos(tag))
    #     for word, tag in tagged_words
    # ]

    lemmatized = []

    for word, tag in tagged_words:
        lemmatized.append(lemmatizer.lemmatize(word, get_wordnet_pos(tag)))
    
    return " ".join(lemmatized)
```

```python
movies["bag"] = movies["bag"].apply(lemmatize_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(movies["bag"])
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()[:1000]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(movie_name):
    movie_name = movie_name.lower()
    
    matches = movies[movies["title"].str.lower().str.contains(movie_name)]
    
    if matches.empty:
        print("Movie not found 😅")
        return
    
    idx = matches.index[0]
    
    # 🔥 KEY LINE (on-the-fly similarity)
    dist = cosine_similarity(vectors[idx], vectors).flatten()
    
    scores = sorted(list(enumerate(dist)), reverse = True, key = lambda x : x[1])[1:6]

    rec = []
    # top 5
    for i in scores:
        rec.append(movies.iloc[i[0]].title)

    return rec
```

```python
recommend("shawshank")
```

```python
recommend("fight club")
```

```python
recommend("toy story")
```
