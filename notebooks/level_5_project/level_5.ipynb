```python
# Constants based on your descriptive stats
m = 34
C = 6.314

def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(v+m) * C)
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def lemmatize_text(text):
    words = text.split()
    tagged_words = pos_tag(words)

    lemmatized = []

    for word, tag in tagged_words:
        lemmatized.append(lemmatizer.lemmatize(word, get_wordnet_pos(tag)))
    
    return " ".join(lemmatized)
```

```python
movies["tag"] = movies["tag"].apply(lemmatize_text)
```

```python
min_votes = movies["vote_count"].min()
max_votes = movies["vote_count"].max()

def norm_votes(v):
    return (v - min_votes) / (max_votes - min_votes)
```

```python
def recommend(movie_name):
    movie_name = movie_name.lower()
    
    # Priority 1: Exact match 
    match = movies[movies["title"].str.lower() == movie_name]
    
    # Priority 2: Contains (If exact match fails)
    if match.empty:
        match = movies[movies["title"].str.lower().str.contains(movie_name)]
    
    if match.empty:
        print("Movie not found 😅")
        return
    
    # Grab the most popular one if multiple matches exist
    idx = match.sort_values('rating', ascending=False).index[0]
    
    # Calculate similarity on-the-fly
    dist = cosine_similarity(vectors[idx], vectors).flatten()
    
    # 1. Get the top 20 most SIMILAR movies first
    # [1:21] skips the movie itself and takes the next 20
    similar_indices = sorted(list(enumerate(dist)), reverse=True, key=lambda x: x[1])[1:51]

    # 2. Extract these 20 movies into a list so we can sort them by rating
    top_20_list = []

    for i in similar_indices:
        idx_i = i[0]
        sim_score = i[1]
        temp_df = movies.iloc[idx_i]
    
        # filter weak movies
        if temp_df['rating'] < 6.5:
            continue
    
        rating_score = temp_df['rating'] / 10
        popularity_score = norm_votes(temp_df['vote_count'])
    
        final_score = (
            0.6 * sim_score +
            0.25 * rating_score +
            0.15 * popularity_score
        )
    
        top_20_list.append({
            'title': temp_df['title'],
            'rating': temp_df['rating'],
            'score': final_score
        })

    # 3. Sort those 20 candidates by RATING (Highest first)
    final_recommendations = sorted(top_20_list, key=lambda x: x['score'], reverse=True)

    print(f"Top 10 High-Quality Recommendations for: {movies.loc[idx, 'title']}")
    print("-" * 30)
    
    # 4. Display and return the top 10 from this sorted list
    rec = []
    for movie in final_recommendations[:10]:
        print(f"- {movie['title']} (Rating: {movie['rating']:.2f})")
        rec.append(movie['title'])

    return rec
```

```python
def director_spotlight_by_idx(target_idx, top_n=5):
    # Get director directly from the known index
    target_director = movies.loc[target_idx, 'crew']
    
    # Filter dataset for other movies by this director
    director_movies = movies[(movies['crew'] == target_director) & (movies.index != target_idx)]
    
    if director_movies.empty:
        return target_director, []

    # Get Top 5 by rating
    top_dir_df = director_movies.sort_values('rating', ascending=False).head(top_n)
    director_recs = top_dir_df[['title', 'rating']].to_dict('records')

    return target_director, director_recs
```

```python
def enhanced_recommend(movie_title):
    """
    Controller function that finds the specific movie index and triggers 
    both the Content-Based Top 10 and the Director's Spotlight.
    """
    # 1. Find the specific movie and its unique index
    # (Matches against your 'Title (Year)' format)
    movie_title_lower = movie_title.lower()
    match = movies[movies["title"].str.lower() == movie_title_lower]

    if match.empty:
        match = movies[movies["title"].str.lower().str.contains(movie_title_lower)]

    if match.empty:
        print("Movie not found 😅")
        return

    # Grab the index of the highest rated match to ensure quality
    target_idx = match.sort_values('rating', ascending=False).index[0]
    final_title = movies.loc[target_idx, 'title']

    print("\n" + "="*60)
    
    # 2. Part A: Run your High-Quality Top 10 Recommender
    # We pass the title found to your existing recommend() logic
    recommend(final_title)

    # 3. Part B: Run the Director's Spotlight
    # We pass the target_idx directly to avoid the 'remake/animation' director swap
    director_name, dir_recs = director_spotlight_by_idx(target_idx)

    if dir_recs:
        # Simple cleanup for the joined string names (e.g., martinscorsese -> Martinscorsese)
        display_name = director_name.capitalize()
        
        # If you want to handle the "MartinScorsese" casing specifically:
        import re
        display_name = re.sub(r"([a-z])([A-Z])", r"\1 \2", display_name).title()

        print(f"\n🎥 Director's Spotlight: More from {display_name}")
        print("-" * 30)
        for d_movie in dir_recs:
            print(f"- {d_movie['title']} (Rating: {d_movie['rating']:.2f})")
    else:
        print(f"\n🎥 No other high-quality films by this director in our database.")
```
